```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
df= pd.read_excel("Gastos y costos 20-23.xlsx",sheet_name= "2023", header=3)
```

Imputar valores nulos

```python
#Identificar valores nulos por columna
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos por dataframe
valores_nulos=df.isnull().sum().sum()
valores_nulos
```

```python
#Se rellenaron lo naN con bffill ya que rellena los valores faltantes
#utilizando el valor más reciente pero después del valor faltante.
df =df.fillna(method="bfill")
df =df.fillna(method="ffill")
df.info()
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
data1 = df.drop(['Otros'], axis=1)

data1.info()
```

```python
data1.info()
```

Valores atípicos

```python
fig = plt.figure(figsize =(15, 8))
data1.plot(kind='box', vert=False)
plt.title("Valores Atípicos de Gastos y costos 20-23")
plt.show()
```

#Procedimiento Desviación Estándar

```python
y=data1
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
data2= data1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2
```

```python
#Eliminamos columnas innecesarias
data3 = data2.drop(['Fecha',"Folio","UUID","RFC","Proveedor",
                   "TIPO GASTO","Descripción","MP","Tipo","Status","Poliza"], axis=1)

data3
```

```python
data3.info()
```

```python
#Corroboramos valores nulos del df_std1
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3.fillna(data3.mean(), inplace=True)
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
fig = plt.figure(figsize =(15, 8))
data3.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

##Rango Intercuartílico

```python
#En este codigo se calcula los límites superior e inferior permitidos para una variable utilizando el IQR (rango) y el factor de 1.5 veces
#el IQR como criterio para identificar valores atípicos potenciales. Los valores que estén
#por encima del límite superior o por debajo del límite inferior se considerarán atípicos en la distribución de la variable
y=data2

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
data4 = data2.copy()
```

```python
data5 = data4.drop(['Fecha',"Folio","UUID","RFC","Proveedor",
                   "TIPO GASTO","Descripción","MP","Tipo","Status","Poliza"], axis=1)
data5
```

```python
valores_nulos=data5.isnull().sum()
valores_nulos
```

```python
data6 = data5.copy()
data6 = data6.fillna(round(data5.mean(),1))
data6
```

```python
fig = plt.figure(figsize =(15, 8))
data6.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

```python
indice=df['Fecha']
Datos_limpiosStd = pd.concat([indice, data5], axis=1)
Datos_limpiosStd
```

```python
indice=df['Fecha']
Datos_limpiosQ = pd.concat([indice, data6], axis=1)
Datos_limpiosQ
```
